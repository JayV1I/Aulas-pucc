```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
```

```python

 #cluster
variaveis = ['SO2', 'temp', 'manu']
df_cluster = df.dropna(subset=variaveis).copy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster[variaveis])
#ward variancia minima cluster
linked = linkage(X_scaled, method='ward')

 # Dendrograma
plt.figure(figsize=(12, 6))
dendrogram(linked, labels=df_cluster['city'].values, leaf_rotation=90)
plt.title("Dendrograma - Clusterização Hierárquica (Ward)")
plt.xlabel("Países")
plt.ylabel("Distância")
plt.tight_layout()
plt.show()
```

```python
# Filtro
print("Unique values in the 'city' column:")
print(df['city'].unique())

paises = [
    'Albany', 'Albuquerque', 'Atlanta', 'Baltimore', 'Birmingham'
]
print(df.columns)

df_sub = df[df['city'].isin(paises)].dropna(subset=variaveis).copy()

if df_sub.empty:
    print("Filtered DataFrame is empty. The selected cities might not exist or have missing values in the specified columns.")
else:

    X_sub_scaled = StandardScaler().fit_transform(df_sub[variaveis])

    # Linkage Ward
    linked_sub = linkage(X_sub_scaled, method='ward')
    plt.figure(figsize=(10, 5))

    dendrogram(linked_sub, labels=df_sub['city'].values, leaf_rotation=90)
    plt.title("Dendrograma - Cidades Selecionadas (Clusterização Hierárquica)")
    plt.xlabel("Cidades")
    plt.ylabel("Distância")
    plt.tight_layout()
    plt.show()
```

```python

X = df.drop(columns=["city"])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
#metodos
methods = ['single', 'complete', 'average', 'ward']

plt.figure(figsize=(16, 12))
for i, method in enumerate(methods, 1):
    plt.subplot(2, 2, i)
    Z = linkage(X_scaled, method=method)
    dendrogram(Z, labels=df["city"].values, leaf_rotation=90)
    plt.title(f"Dendrograma - Método: {method}")

plt.tight_layout()
plt.show()
```

O método Ward apresentou os grupos mais bem definidos, com separações mais claras entre os ramos, sem cadeias encadeadas ou fusões prematuras como no método Single.
