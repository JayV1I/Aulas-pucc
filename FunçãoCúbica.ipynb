# USO NO COTIDIANO DE FUNES CBICAS:

>Animaes e Design Grfico: Na computao grfica, funes cbicas so usadas para suavizar transies e curvas em animaes, criando trajetrias mais naturais para objetos em movimento.

>Exemplo da funo aplicada em IA. anti alising:Screenshot-2021-10-18-at-19-20-39-anti-aliasing.webpanti.gif

```python


# Funo cbica com 3 razes, por exemplo: f(x) = (x - 1)(x + 2)(x - 3) => f(x) = x^3 - 2x^2 - 5x + 6
def f(x):
    return (x - 1)*(x + 2)*(x - 3)

# Gerando valores de x para o grfico
x = np.linspace(-4, 4, 400)
y = f(x)

# Criando o grfico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = (x - 1)(x + 2)(x - 3)', color='b')
plt.axhline(0, color='black',linewidth=1)  # eixo x
plt.axvline(0, color='black',linewidth=1)  # eixo y
plt.title("Grfico de uma funo cbica:")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Criando o grfico sem a linha e com pontos nas intersees
plt.figure(figsize=(8, 6))
plt.scatter([1, -2, 3], [0, 0, 0], color='r', label='Interseces no eixo X', zorder=5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.title("Grfico de uma funo cbica com interseces no eixo X")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python

# Definindo a funo cbica a partir das razes e dos coeficientes a e d
def plot_cubic_function(a, r1, r2, r3, d):
    # Coeficientes da funo cbica a partir das razes e do coeficiente a
    b = -a * (r1 + r2 + r3)
    c = a * (r1 * r2 + r2 * r3 + r1 * r3)

    # Gerando a funo cbica com os parmetros dados
    def func(x):
        return a * x**3 + b * x**2 + c * x + d

    # Criando um intervalo de valores para x
    x_vals = np.linspace(-10, 10, 400)
    y_vals = func(x_vals)

    # Plotando o grfico
    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label=f'f(x) = {a:.1f}x + {b:.1f}x + {c:.1f}x + {d:.1f}', color='blue', linewidth=2)

    # Ajustando o estilo do plano cartesiano
    plt.axhline(0, color='black', linewidth=1)  # Eixo x
    plt.axvline(0, color='black', linewidth=1)  # Eixo y
    plt.xticks(np.arange(-10, 11, 1))  # Marcas no eixo x de -10 a 10 com passo de 1
    plt.yticks(np.arange(-100, 101, 10))  # Marcas no eixo y de -100 a 100 com passo de 10
    plt.xlim(-10, 10)
    plt.ylim(-100, 100)
    plt.grid(True, linestyle='--', alpha=0.6, color='gray')  # Grade mais visvel

    # Calculando as razes reais da funo
    roots = np.roots([a, b, c, d])
    real_roots = [r.real for r in roots if np.isreal(r)]  # Filtrando apenas as razes reais

    # Marcando as razes reais no grfico
    plt.scatter(real_roots, [0]*len(real_roots), color='red', zorder=5)

    # Adicionando coordenadas das razes reais
    for r in real_roots:
      plt.text(r, 0, f'({r:.0f}, 0)', fontsize=10, verticalalignment='bottom')

    # Ajustando ttulo e legenda
    plt.title('Grfico Interativo da Funo Cbica', fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('f(x)', fontsize=12)
    plt.legend()
    plt.show()

# Criando sliders interativos para os coeficientes e as razes da funo cbica
interact(plot_cubic_function,
         a=IntSlider(min=-10, max=10, step=1, value=1, description='Coeficiente a'),
         r1=IntSlider(min=-10, max=10, step=1, value=1, description='Raiz 1'),
         r2=IntSlider(min=-10, max=10, step=1, value=4, description='Raiz 2'),
         r3=IntSlider(min=-10, max=10, step=1, value=7, description='Raiz 3'),
         d=IntSlider(min=-10, max=10, step=1, value=0, description='Coeficiente d'))
```

# Coeficiente 
a: Controla a **intensidade** da curvatura cbica. Alterar esse valor afeta diretamente a inclinao da funo e a maneira como ela sobe ou desce.

# Coeficiente 
b: Controla a **concavidade** e a posio da crista. Mudando o valor de

b, voc pode **alterar onde o ponto mximo ou mnimo local** aparece e quo "aberto" ou "fechado" o ponto est.

# Coeficientes c e 
d: Embora esses coeficientes tenham um efeito menos direto na crista,

c **ajusta a inclinao linear** e

d **desloca** **verticalmente** o grfico
