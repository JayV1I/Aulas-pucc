```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider
```

A definição de uma função cúbica é:
  um produto polinomial de terceiro grau.
# Tendo a sua forma geral, f(x) = ax³ + bx² + cx + d

# USO NO COTIDIANO DE FUNÇÕES CÚBICAS:

>Animações e Design Gráfico: Na computação gráfica, funções cúbicas são usadas para suavizar transições e curvas em animações, criando trajetórias mais naturais para objetos em movimento.

>Exemplo da função aplicada em IA. anti alising:Screenshot-2021-10-18-at-19-20-39-anti-aliasing.webpanti.gif

```python


# Função cúbica com 3 raízes, por exemplo: f(x) = (x - 1)(x + 2)(x - 3) => f(x) = x^3 - 2x^2 - 5x + 6
def f(x):
    return (x - 1)*(x + 2)*(x - 3)

# Gerando valores de x para o gráfico
x = np.linspace(-4, 4, 400)
y = f(x)

# Criando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = (x - 1)(x + 2)(x - 3)', color='b')
plt.axhline(0, color='black',linewidth=1)  # eixo x
plt.axvline(0, color='black',linewidth=1)  # eixo y
plt.title("Gráfico de uma função cúbica:")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Criando o gráfico sem a linha e com pontos nas interseções
plt.figure(figsize=(8, 6))
plt.scatter([1, -2, 3], [0, 0, 0], color='r', label='Intersecções no eixo X', zorder=5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.title("Gráfico de uma função cúbica com intersecções no eixo X")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python

# Definindo a função cúbica a partir das raízes e dos coeficientes a e d
def plot_cubic_function(a, r1, r2, r3, d):
    # Coeficientes da função cúbica a partir das raízes e do coeficiente a
    b = -a * (r1 + r2 + r3)
    c = a * (r1 * r2 + r2 * r3 + r1 * r3)

    # Gerando a função cúbica com os parâmetros dados
    def func(x):
        return a * x**3 + b * x**2 + c * x + d

    # Criando um intervalo de valores para x
    x_vals = np.linspace(-10, 10, 400)
    y_vals = func(x_vals)

    # Plotando o gráfico
    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label=f'f(x) = {a:.1f}x³ + {b:.1f}x² + {c:.1f}x + {d:.1f}', color='blue', linewidth=2)

    # Ajustando o estilo do plano cartesiano
    plt.axhline(0, color='black', linewidth=1)  # Eixo x
    plt.axvline(0, color='black', linewidth=1)  # Eixo y
    plt.xticks(np.arange(-10, 11, 1))  # Marcas no eixo x de -10 a 10 com passo de 1
    plt.yticks(np.arange(-100, 101, 10))  # Marcas no eixo y de -100 a 100 com passo de 10
    plt.xlim(-10, 10)
    plt.ylim(-100, 100)
    plt.grid(True, linestyle='--', alpha=0.6, color='gray')  # Grade mais visível

    # Calculando as raízes reais da função
    roots = np.roots([a, b, c, d])
    real_roots = [r.real for r in roots if np.isreal(r)]  # Filtrando apenas as raízes reais

    # Marcando as raízes reais no gráfico
    plt.scatter(real_roots, [0]*len(real_roots), color='red', zorder=5)

    # Adicionando coordenadas das raízes reais
    for r in real_roots:
      plt.text(r, 0, f'({r:.0f}, 0)', fontsize=10, verticalalignment='bottom')

    # Ajustando título e legenda
    plt.title('Gráfico Interativo da Função Cúbica', fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('f(x)', fontsize=12)
    plt.legend()
    plt.show()

# Criando sliders interativos para os coeficientes e as raízes da função cúbica
interact(plot_cubic_function,
         a=IntSlider(min=-10, max=10, step=1, value=1, description='Coeficiente a'),
         r1=IntSlider(min=-10, max=10, step=1, value=1, description='Raiz 1'),
         r2=IntSlider(min=-10, max=10, step=1, value=4, description='Raiz 2'),
         r3=IntSlider(min=-10, max=10, step=1, value=7, description='Raiz 3'),
         d=IntSlider(min=-10, max=10, step=1, value=0, description='Coeficiente d'))
```

# Coeficiente 𝑎
a: Controla a **intensidade** da curvatura cúbica. Alterar esse valor afeta diretamente a inclinação da função e a maneira como ela sobe ou desce.

# Coeficiente 𝑏
b: Controla a **concavidade** e a posição da crista. Mudando o valor de
𝑏
b, você pode **alterar onde o ponto máximo ou mínimo local** aparece e quão "aberto" ou "fechado" o ponto está.

# Coeficientes c e 𝑑
d: Embora esses coeficientes tenham um efeito menos direto na crista,
𝑐
c **ajusta a inclinação linear** e
𝑑
d **desloca** **verticalmente** o gráfico
